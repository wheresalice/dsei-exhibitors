```python
import pandas as pd
```

```python
df_2021 = pd.read_parquet("../parquet/dsei-2021")
df_2019 = pd.read_parquet("../parquet/dsei-2019")
df_2017 = pd.read_parquet("../parquet/dsei-2017")
df_2015 = pd.read_parquet("../parquet/dsei-2015")
df_2013 = pd.read_parquet("../parquet/dsei-2013")
df_2011 = pd.read_parquet("../parquet/dsei-2011")
df_2009 = pd.read_parquet("../parquet/dsei-2009")
df_2007 = pd.read_parquet("../parquet/dsei-2007")
df_2005 = pd.read_parquet("../parquet/dsei-2005")
df_2003 = pd.read_parquet("../parquet/dsei-2003")
df_2001 = pd.read_parquet("../parquet/dsei-2001")
df_1999 = pd.read_parquet("../parquet/dsei-1999")
```

```python
df = pd.concat([df_2021, df_2019, df_2017, df_2015, df_2013, df_2011, df_2009, df_2007, df_2005, df_2003, df_2001, df_1999], axis=1)
```

```python
import leafmap
import geopandas

m = leafmap.Map()

map_2021 = df_2021.astype({'address_lon': float, 'address_lat': float})

gdf = geopandas.GeoDataFrame(map_2021, geometry=geopandas.points_from_xy(map_2021.address_lon, map_2021.address_lat))
gdf.crs="EPSG:4326"
m.add_gdf(gdf, layer_name="DSEI 2021 Exhibitors", info_mode='on_click')
m.zoom_to_bounds((-10, 49, 2, 60))
m
```
